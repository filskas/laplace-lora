```python
import json
```

```python
import pandas as pd
```

```python
ds = []
for l in open("eval_res.json", "r").readlines():
    if l.strip() == "":
        continue
    try:
        d = json.loads(l)
        ds.append(d)
    except json.JSONDecodeError:
        print("Error decoding JSON:", l)
        continue

    # Convert the JSON object to a DataFrame
```

```python
df = pd.DataFrame(ds)
```

```python
import torch
```

```python
probs = []
for logits in df["logits"]:
    logits = torch.tensor(logits)
    probs1 = torch.softmax(logits, dim=-1)
    probs.append(probs1)
probs = torch.stack(probs)
```
